# Graphical User Interface

To run SOFA with a GUI, we need two modules:

## Sofa.GUI

```python
import Sofa.Gui
```

## SofaImGui

If not found, the path to this module must be added to `PYTHONPATH`.

```python
import SofaImGui
```

## Empty Scene

Let's define again a empty graph scene:

```python
root = Sofa.Core.Node("root")
```

```python
root.addObject("DefaultAnimationLoop")
```

Initialize the root node:

```python
Sofa.Simulation.initRoot(root)
```

## GUI

Initialize the GUI. It will open a window, but the GUI does not run.

```python
Sofa.Gui.GUIManager.Init("myscene", "imgui")
Sofa.Gui.GUIManager.createGUI(root, "Jupyter notebook file")
Sofa.Gui.GUIManager.SetDimension(1080, 800)
```

Run the GUI. The window won't be closed automatically.

```python
Sofa.Gui.GUIManager.MainLoop(root)
Sofa.Gui.GUIManager.closeGUI()
```

If you want to change something in the scene, you need to first close the window:

```python
root.addObject("RequiredPlugin", name="Sofa.Component.StateContainer")
```
