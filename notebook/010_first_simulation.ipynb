# Run you first simulation with SOFA

1. Import the SOFA module:

```python
import Sofa
```

2. Call the SOFA function to create the root node. At this stage, the scene graph is empty.

```python
root = Sofa.Core.Node("root")
```

3. Initialize the root node

```python
Sofa.Simulation.initRoot(root)
```

3. Run 10 simulation steps

```python
# Run 10 simulation steps
for iteration in range(10):
    Sofa.Simulation.animate(root, root.dt.value)
    print(f"Computing iteration {iteration}")

print("Computation is done.")
```
